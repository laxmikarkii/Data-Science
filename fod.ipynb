```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/Users/laxmii/Documents/Data Science Lab/Lab/Health_heart_experimental.csv")
df.head()
```

## Line chart

```python
age = list(range(0, 20))
height = [50, 75, 85, 95, 105, 115, 125, 135, 145, 150,
          155, 160, 165, 170, 172, 174, 175, 176, 177, 178]

plt.figure()
plt.plot(age, height)
plt.xlabel("Age (years)")
plt.ylabel("Height (cm)")
plt.title("Increase in Height with Age")
plt.grid(True)
plt.show()
```

In this graph, the line is most steep between 0 to 1, that means a child grows most rapidly from birth to age 1. Also, the height of a child grows most consistently from age 1 to 8. And the growth starts to slow down as adlolescent phase starts and stops till the age of 18 completely.

## Bar Plot

```python
avg_height = df.groupby("sex")["heightCm"].mean()

plt.figure()
plt.bar(["Female", "Male"], avg_height)
plt.xlabel("Gender")
plt.ylabel("Average Height (cm)")
plt.title("Average Height by Gender")
plt.show()
```

Here, female have an average height, around 155-158cm and male have an average height only sightly more than female, around 158-160cm. The difference doesn`t seem much, this could because this is average value which can cluster closely and not reflect general population.

## Box Plot

## Histogram

```python
plt.figure()
plt.hist(df["BMI"], bins=7)
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.title("BMI Distribution")
plt.show()
```

Here the histogram shows that most people have a BMI between 25 and 35, indicating a prevalence of overweight and obesity in the dataset. Extreme values are less frequent which suggests that the data is moderately spread without major outliers.

## Linear Algebra Task

```python
# matrix operation
import numpy as np
A=np.array([[2,-1],
             [0,3]])
B=np.array([[1,2],
             [-3,4]])
print("Matrix A:", A)
print(A.shape)
print("Matrix B:", B)
print(B.shape)
```

```python
# a. A+B
C = A + B
print("Matrix A + Matrix B =\n", C)
print(C.shape)
```

```python
# b. AB
print("Dot product using 'np.dot' --->",np.dot(A, B)) #

print("Dot product using '.dot' --->",A.dot(B))

print("Dot product using '@' --->",A @ B)
print((A @ B).shape)
```

```python
# c. 3A-2B
alpha = 3
beta = 2
D = alpha * A - beta * B
print("3A - 2B =\n", D)
print(D.shape)
```

```python
# 2 Determinant of A
import numpy.linalg as LA
LA.det(A)  
```

```python
# Since det(A)  0, matrix A is invertible.
LA.inv(A)  # inverse of A
```

```python
# Eigen values and Eigen vectors
C = np.array([[ 4,  1],
              [-2,  1]])

eigenvalues, eigenvectors = np.linalg.eig(C)

print("Matrix C:")
print(C)
print("\nEigenvalue (first):")
print(eigenvalues[0])
print("\nCorresponding Eigenvector:")
print(eigenvectors[:, 0])
```

```python
# 4predicting price
X = np.array([
    [1, 4, 1.5, 100],
    [1, 6, 2.0, 150]
])

# Price vector
y = np.array([20, 30])

# MoorePenrose Pseudo Inverse
beta = np.linalg.pinv(X) @ y
```

```python
car3 = np.array([1, 8, 3.0, 200])
price_car3 = car3 @ beta
price_car3
print("Predicted price for car3:", price_car3)
```
